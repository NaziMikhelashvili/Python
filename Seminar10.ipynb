```python
penguins = sns.load_dataset('penguins')
penguins.info()
```

```python
sns.scatterplot(data =penguins, x='bill_depth_mm', y='body_mass_g', palette = 'gray', hue = 'sex', size = 'species')
```

```python
sns.set(style='ticks', context = 'paper', palette ='rainbow')
g = sns.PairGrid(penguins, diag_sharey = False, hue = 'sex')
g.map(sns.scatterplot)
```

```python
sns.histplot(data=penguins, x='island', hue ='sex', palette ='inferno')
```

```python
import numpy as np
lst = list(penguins['bill_length_mm'])[:324]
lst = np.array(lst).reshape((18,18))
sns.heatmap(data = lst, annot = True)
```

## Задача
создать столбец который бдет отражать длину пингвина - high medium low

```python
penguins.loc[penguins['bill_length_mm']>=42, 'height_group'] = 'High'
penguins.loc[(penguins['bill_length_mm']<42)& (penguins['bill_length_mm']>=35), 'height_group'] = 'Middle'
penguins.loc[penguins['bill_length_mm']<35, 'height_group'] = 'Low'
```

```python
sns.histplot(data=penguins, x='height_group', hue ='sex')
```
